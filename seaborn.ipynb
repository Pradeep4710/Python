```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('tips')
df
```

```python
df['total_bill'].max()
```

```python
df['total_bill'].min()
```

```python
sns.set()
```

```python
# Distribution plot (Histograph)
sns.displot(df['total_bill'],kde=True)
#KDE --> Kernel Density Estimation
```

```python
sns.scatterplot(x = 'total_bill' , y = 'tip' , data = df , color = 'Green')
```

```python
sns.relplot(x='total_bill',y='day',data=df)
```

```python
#bivariate analysis
sns.jointplot(x='total_bill',y='tip',data=df)
```

```python
sns.pairplot(df)
```

```python
# Box plot helps us with the 5 number summary
# 0th percentile
# 25th percentile
# 50th percentile
# 75th percentile
# 100th percentile

sns.boxplot(x=df['total_bill'])
```

```python
sns.catplot(x='day',y='total_bill',data=df)
# by default graph is set on stripplot
```

```python
sns.catplot(x='day',y='total_bill',data=df,kind='box')
```

```python
sns.catplot(x='day',y='total_bill',data=df,kind='strip')
```

```python
sns.catplot(x='day',y='total_bill',data=df,kind='boxen')
```
